```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(y,x,dx): 
    return (y(x+dx)-y(x))/dx

x=1.0 
dx=0.1 
print('x=',x,' and ', 'dx=', dx)
print('forward difference',f(np.cos,x,dx))
```

```python
def F(name, n, fwd, cntl):
    f = open(name, "w")
    for i in range(len(n)):
        f.write(str(n[i]))
        f.write("\t")
        f.write(str(fwd[i]))
        f.write("\t")
        f.write(str(cntl[i]))
        f.write("\t")
    f.close()
    return


#forward
def derifw(y,x,dx): 
    return (y(x+dx)-y(x))/dx


# central 
def dericd(y,x,dx): 
    return (y(x+dx/2)-y(x-dx/2))/dx

x= np.pi/4
dxarr=[]
yarr_dericd=[]
yarr_derifw=[]

for dx in 10**np.arange(-16, 0.1,0.01):
    y=dericd(np.cos,x,dx) 
    dxarr.append(dx) 
    yarr_dericd.append(y) 
    y=derifw(np.cos,x,dx) 
    yarr_derifw.append(y)
    
F("HW3 data.txt", dxarr, yarr_dericd, yarr_derifw)  
print(yarr_derifw)
```

```python
plt.plot(dxarr, abs(yarr_derifw-np.sin(np.pi/4)),label='forward')     
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.legend()
plt.show()

print('n=0 and dx=10^15')
```

```python
plt.plot(dxarr,abs(yarr_dericd-np.sin(np.pi/4)),'r-',lw=2,label='central')
plt.plot(dxarr, abs(yarr_derifw-np.sin(np.pi/4)),'b-', lw=2, label='forward')     
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.legend(loc= 'lower right')
plt.show()
print('CD is better as the error is smaller by a factor of nearly 10')
```
