```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
from sympy import *
```

```python
x=symbols("x")
```

```python
i=1
n=11
sum=0
while(i<n+1):
    sum=sum+sin(i*x)/i
    i=i+2
#print(sum)
```

```python
g=integrate(sum,x)
print(g)
```

```python
x=-np.pi
a=-np.cos(x)-np.cos(3*x)/9-np.cos(5*x)/25-np.cos(7*x)/49-np.cos(9*x)/81-np.cos(11*x)/121
b=a
```

```python
x=np.arange(-np.pi,np.pi,0.01)
m=-np.cos(x)-np.cos(3*x)/9-np.cos(5*x)/25-np.cos(7*x)/49-np.cos(9*x)/81-np.cos(11*x)/121-b

plt.figure(figsize=(10,8))
ax=pl.gca()
plt.xlabel(r"$x$",fontsize=27)
plt.ylabel(r"$m(x)$",fontsize=27)
plt.tick_params(labelsize=20)

plt.plot(x,m)
```
