```python
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
from numpy import *
import numpy as np
Nmax = 100; Niter = 500; V = zeros((Nmax, Nmax), float)   
V[50,50] = 1.0*1.0*4*np.pi
i_iter = np.array(range(1,Nmax-1))
j_iter = i_iter
print(V[50,50])
for iter in range(Niter):                                  
    if iter%100 == 0: print(iter)
    maxerr=0.
    for i in i_iter:
        for j in j_iter: 
            if i == j == 50:
                break 
            du = 1./6.*(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1]+2*V[i,j])-V[i,j]
            if(maxerr<abs(du)): 
                maxerr=abs(du)  
            V[i,j]= V[i,j]+du

    if iter%100 ==0:print(maxerr)
    if(maxerr<1.e-3):
        break      
x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X,Y=plt.meshgrid(x,y)                 
def functz(V):                                       
    z = V[X,Y]                        
    return z
Z = functz(V)                          
fig = plt.figure()                                      
ax = Axes3D(fig)                                     
ax.plot_wireframe(X, Y, Z, color = 'r')              
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
plt.show()                                              

x_plot = np.arange(50)
y_plot = V[50,:50]
print(y_plot)
plt.plot(x_plot,y_plot,c="black")

x_test=np.linspace(0.0001,50,500)
y_test=-2*1*1*np.log(-x_test+50)+8

plt.plot(x_test,y_test,c="red")
```
