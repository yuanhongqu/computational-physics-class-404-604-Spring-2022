```python
import time 
import numpy as np 
import pylab as pl
import matplotlib.pyplot as plt
```

```python
def fib_memo(n, table):
    if n == 0 or n == 1:  
        table[n]= n     
    if table[n] is None:  
        table[n] = fib_memo(n-1,table) + fib_memo(n-2,table) 
    return table[n]

def fib_ratio(n): 
    table = [None]*(n+1) 
    table[0]= 0  
    fib_num = fib_memo(n,table)
    fib_denom = table[n-1] 
    return fib_num/fib_denom

x_value = np.arange(1,11)
y_value = [] 
```

```python
for x in x_value:
    y_value.append(fib_ratio(x))
plt.figure(figsize=(10,8))
ax=pl.gca()
plt.tick_params(labelsize=20)
plt.plot(x_value,y_value)
plt.show()
```
