```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
x=np.arange(-np.pi,np.pi,0.01)
```

```python
i=1
j=0
k=0
n_1=11
n_2=21
n_3=31
sum_1=0
sum_2=0
sum_3=0
while(i<n_1+1):
    sum_1=sum_1+np.sin(i*x/i)
    i=i+2
    
while(j<n_2+1):
    sum_2=sum_2+np.sin(i*x/i)
    j=j+2
    
while(k<n_3+1):
    sum_3=sum_3+np.sin(i*x/i)
    k=k+2
```

```python
f_1=np.pi/2+2*sum_1
f_2=np.pi/2+2*sum_2
f_3=np.pi/2+2*sum_3
```

```python
plt.figure(figsize=(10,8))
ax=pl.gca()
plt.xlabel(r"$x$",fontsize=27)
plt.ylabel(r"$f(x)$",fontsize=27)
plt.tick_params(labelsize=20)
plt.plot(x,f_1,c="black",label="n=11")
plt.plot(x,f_2,c="red",label="n=21")
plt.plot(x,f_3,c="deepskyblue",label="n=31")
plt.legend(loc='upper left',fontsize=20)
```
